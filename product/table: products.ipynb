```python
import sys
import json
import os
```

```python
parent_dir = os.path.abspath('..') 
if parent_dir not in sys.path:
    print(f'Adding {parent_dir} to path')
    sys.path.append(parent_dir) # Adjust path as needed

import createjsonschema
import createtablefromjsonschema
```

```python
jsonschemaname = 'products'
domain = 'tables'
jsonschemadescription = 'This is a jsonschema file defining the table for validating a product record'

objectsting = ''

objectsting = createjsonschema.schema(jsonschemaname,objectsting).create()
objectsting = createjsonschema.schema(jsonschemaname,objectsting).description(jsonschemadescription).modify()
objectsting = createjsonschema.schema(jsonschemaname,objectsting).catalog("mdm").add()
objectsting = createjsonschema.schema(jsonschemaname,objectsting).schema("product").add()
objectsting = createjsonschema.schema(jsonschemaname,objectsting).schema("product").add()
objectsting = createjsonschema.schema(jsonschemaname,objectsting).properties("productkey", "The unique identifier for the product", "biginterger").add()
objectsting = createjsonschema.schema(jsonschemaname,objectsting).properties("vendorname", "The name of the vendor", "string").add()
objectsting = createjsonschema.schema(jsonschemaname,objectsting).properties("catalogid", "the id assigned by a vendor in a catalog", "string").add()
objectsting = createjsonschema.schema(jsonschemaname,objectsting).properties("productdescription", "The description of the product", "string").add()
objectsting = createjsonschema.schema(jsonschemaname,objectsting).required(["vendorname"]).add()
objectsting = createjsonschema.schema(jsonschemaname,objectsting).required(["catalogid"]).add()
objectsting = createjsonschema.schema(jsonschemaname,objectsting).required(["productdescription"]).add()
objectsting = createjsonschema.schema(jsonschemaname,objectsting).additionalproperties({"primarykeycolumn": "productkey"}).modify()
objectsting = createjsonschema.schema(jsonschemaname,objectsting).additionalproperties({"uniquecolumns": ["vendorname","catalogid","productdescription"]}).modify()
objectsting = createjsonschema.schema(jsonschemaname,objectsting).tags("datadomain", "products").add()

print(json.dumps(json.loads(objectsting), indent=4))
```

```python
sql = createtablefromjsonschema.JsonSchemaTableCreator(objectsting).create_table()


print(sql)
```

```python
%sql

DROP TABLE IF EXISTS mdm.product.products;

CREATE TABLE mdm.product.products (
  productkey BIGINT
   GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1)
  ,vendorname STRING NOT NULL
  ,catalogid STRING NOT NULL
  ,productdescription STRING NOT NULL
  ,CONSTRAINT products_pk PRIMARY KEY (productkey)
  --,CONSTRAINT products_unique UNIQUE (vendorname, catalogid, productdescription)
)
USING DELTA;

ALTER TABLE mdm.product.products SET TAGS ('datadomain' = 'products');



```
