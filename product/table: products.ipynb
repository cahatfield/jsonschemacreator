```python
import sys
import json
import os
```

```python
parent_dir = os.path.abspath('..') 
if parent_dir not in sys.path:
    print(f'Adding {parent_dir} to path')
    sys.path.append(parent_dir) # Adjust path as needed

import jsonschema
```

```python
jsonschemaname = 'products'
domain = 'tables'
jsonschemadescription = 'This is a jsonschema file defining the table for validating a product record'

filepath = f'./tables/{jsonschemaname}.schema.json'

if os.path.exists(filepath):
    os.remove(filepath)

def print_schema():
     with open(filepath, 'r') as f:
         catalog_schema = json.load(f)
         print(json.dumps(catalog_schema, indent=2))    

jsonschema.schema(jsonschemaname,domain).create()
jsonschema.schema(jsonschemaname,domain).description(jsonschemadescription).modify()
jsonschema.schema(jsonschemaname,domain).properties("productkey", "The unique identifier for the product", "number").add()
jsonschema.schema(jsonschemaname,domain).properties("vendorname", "The name of the vendor", "string").add()
jsonschema.schema(jsonschemaname,domain).properties("catalogid", "the id assigned by a vendor in a catalog", "string").add()
jsonschema.schema(jsonschemaname,domain).properties("productdescription", "The description of the product", "string").add()
jsonschema.schema(jsonschemaname,domain).required(["vendorname"]).add()
jsonschema.schema(jsonschemaname,domain).required(["catalogid"]).add()
jsonschema.schema(jsonschemaname,domain).required(["productdescription"]).add()
jsonschema.schema(jsonschemaname,domain).additionalproperties({"primarykeycolumn": "productkey"}).modify()
jsonschema.schema(jsonschemaname,domain).additionalproperties({"uniquecolumns": ["vendorname","catalogid","productdescription"]}).modify()
# print_schema()
```
